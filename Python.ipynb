## Correlation Plot

```python
df.corr()
```

```python
print(df.corr())

plt.figure(figsize=(10,5))
  
# plotting correlation heatmap
dataplot = sns.heatmap(df.corr(), cmap="YlGnBu", vmin=0, vmax=1)
  
# displaying heatmap
plt.show()
```

## Boxplot 

```python
plt.figure(figsize=(15,8))

# Creating plot
df.boxplot(grid='false', color='blue',fontsize=10, rot=30)
 
# show plot
plt.show()
```

## Bargraph Yoga vs YogaDuration

```python
x= df.Yoga
y = df.YogaDuration

plt.figure(figsize=(15,12))

# Creating Barplot
plt.barh(x, y, height=0.8)
plt.xlabel("Yoga Duration")
plt.ylabel("Yoga")
plt.title("Yoga-YogaDuration")

# show barplot
plt.show()
```

# Bargraph Nakshatra vs NakshatraDuration

```python
x= df.Nakshatra
y = df.NakshatraDuration

plt.figure(figsize=(20,15))

# Creating Barplot
plt.barh(x, y)
plt.xlabel("Nakshatra Duration")
plt.ylabel("Nakshatra")
plt.title("Nakshatra-NakshatraDuration")

# show barplot
plt.show()
```
